```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r'C:\Users\sathvik\Desktop\Preprocessed.csv')
data.head()
```

```python
data = data.drop(['Date'],axis=1)#so i forgot to drop the date..lets do it now
```

```python
data.head(10)
```

```python
data['Absenteeism Time in Hours'].median()
```

```python
#People wh are absent less than 3hrs are moderatly absent and above 3hrs are excessively absent
```

```python
targets = np.where(data['Absenteeism Time in Hours']>3,1,0)
```

```python
targets
```

```python
data['Excessive Absenteeism']=targets
```

```python
data.head(2)
```

```python
targets.sum()/len(targets)
```

```python
#so there are 45% of Excessive Absentees(1s) and 55% moderate(0s) which well balanced to proceed
```

```python
data = data.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data=data.drop(['Daily Work Load Average'],axis=1) #to increase acurracy
```

```python
unscaled_inputs = data.iloc[:,:-1]#taking inputs except last coloumn i.e. Excessive Absenteeism
```

# standardization of data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(unscaled_inputs)
```

```python
scaled_inputs=scaler.transform(unscaled_inputs)
```

# test train split

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,test_size=0.9,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)#to check the accuracy directly instead of manual process i.e. predict
```

```python
#SO OUR ACCURACY ID 78%
```

# finding coefficients & intercepts

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values#there is a mistake. so changing column name reason_3.1 as reason_4
```

```python
unscaled_inputs=unscaled_inputs.rename(columns={'Reason_3.1':'Reason_4'})
```

```python
unscaled_inputs.head()
```

```python
feature_name=unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature name'], data=feature_name)
summary_table['Coefficiants']=np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index=summary_table.index+1
summary_table.loc[0]=['Intercept',reg.intercept_[0]]
summary_table
```

```python
summary_table=summary_table.sort_index()
summary_table
```

```python
summary_table['odds ratio'] = np.exp(summary_table.Coefficiants)
#since it is a log. regression we will apply exp. to coeffiients and see its values 
summary_table
```

```python
summary_table.sort_values('odds ratio',ascending=False)#because by default the order is ascending
```

# testing the data

```python
reg.score(x_test,y_test)
#we removed Daily Work Load Average because it's odds ratio is almost 1 and could have no effect in prediction and increased 
#our accuracy from 69 to 71
```
