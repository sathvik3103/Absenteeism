```python
reasons_coloumn=pd.get_dummies(df['Reason for Absence'],drop_first=True)#to avoid multicollinearity
reasons_coloumn
```

```python
df = df.drop(['Reason for Absence'],axis=1)
df
```

```python
#now before appending the coloumn_reasons to df lets regroup them to avoid adding huge no.of coloumns to our data
#regrouping is based on the type of disease given in table
#s0, 1-14(group1), 
```

```python
reason_type_1 = reasons_coloumn.loc[:, 1:14].max(axis=1)
reason_type_2 = reasons_coloumn.loc[:, 15:17].max(axis=1)
reason_type_3 = reasons_coloumn.loc[:, 18:21].max(axis=1)
reason_type_4 = reasons_coloumn.loc[:, 22:].max(axis=1)
```

```python
df=pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df
```
